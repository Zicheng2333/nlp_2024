```python
#问题4.1
#from bytepiece import Trainer
class corpus:
    def __iter__(self):
        f = 'data.txt'  # 修改文件名为txt
        with open(f, 'r', encoding='utf-8') as f:
            content = f.read()  # 读取整个文件内容
            for item in content.split('\n\n'):# 按双换行符分割
                a=item.strip()
                if(a!=''):
                    yield a  # 每次返回去除首尾空白字符的Unicode字符串
```

```python
f='data.txt'
with open(f,'r',encoding='utf-8') as f:
    content=f.read()
    list=content.split('\n\n')
list
```

```python
trainer = Trainer(order=6, max_vocab_size=600, min_count=32)
trainer.train(corpus(), workers=1, batch_size=1000)
trainer.save('bytepiece2.model')
```

```python
from bytepiece import Tokenizer

tokenizer = Tokenizer('bytepiece2.model')
text = '今天天气不错'

tokens = tokenizer.tokenize(text)  # 返回bytes的list
print(b' '.join(tokens).decode(errors='ignore'))  # 可视化分词结果

ids = tokenizer.encode(text)  # 返回tokens对应的ids
print(tokenizer.decode(ids))  # 重新将ids解码为unicode文本
ids = tokenizer.encode(text, iter=True)  # 返回ids的generator

tokens = tokenizer.tokenize(text, alpha=0.2)  # 随机tokenize
print(b' '.join(tokens).decode(errors='ignore'))  # 可视化分词结果
```
